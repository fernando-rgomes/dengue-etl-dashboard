Criando função de conversão para a coluna idade

```python
def converter_idade(idade):

    if pd.isna(idade):
        return None

    idade = int(idade)

    unidade = idade // 1000
    valor = idade % 1000

    if unidade == 1:   # horas
        return valor / (24 * 365)

    elif unidade == 2: # dias
        return valor / 365

    elif unidade == 3: # meses
        return valor / 12

    elif unidade == 4: # anos
        return valor

    else:
        return None
```

```python
df["idade_anos"] = df["idade"].apply(converter_idade)
```

Criando faixas etárias

```python
df["faixa_etaria"] = pd.cut(
    df["idade_anos"],
    bins=[0, 9, 19, 39, 59, 120],
    labels=[
        "0-9 (crianças)",
        "10-19 (adolescentes)",
        "20-39 (adultos jovens)",
        "40-59 (adultos)",
        "60+ (idosos)"
    ]
)
```

##Limpeza e padronização de variáveis

Mudando variável hospitalização

```python
map_hospitalizacao = {
    1: "Sim",
    2: "Não",
    9: "Ignorado"
}

df["hospitalizacao"] = df["hospitalizacao"].map(map_hospitalizacao)
```

```python
df["hospitalizacao"].value_counts()
```

Remover ignorados

```python
df_hosp = df[df["hospitalizacao"] != "Ignorado"]
```

Criando variável booleana para simplificar a análise

```python
df_hosp["hospitalizado"] = df_hosp["hospitalizacao"] == "Sim"
```

```python
taxa_hosp = (
    df_hosp
    .groupby("faixa_etaria")["hospitalizado"]
    .mean()
) * 100
```

```python
taxa_hosp = taxa_hosp.sort_index()
taxa_hosp
```

```python
import matplotlib.pyplot as plt

taxa_hosp.plot(kind="bar")

plt.title("Taxa de hospitalização por faixa etária")
plt.ylabel("Taxa de hospitalização (%)")
plt.xlabel("Faixa etária")

plt.xticks(rotation=45)

plt.show()
```

idosos têm maior taxa de hospitalização (~10%)

crianças também têm taxa relativamente alta (~7%)

adultos jovens têm menor taxa (~3%)
